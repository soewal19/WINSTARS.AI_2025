```python
Computer Vision CLI Demo
This notebook demonstrates how to use the computer_vision.py file as a CLI (Command Line Interface) command.
The computer_vision.py has two commands train and predict.

Notes:

Logs for the commands are writen into the ./logs directory.
Train Model
Example of training a model using the train command:

python "/path/to/computer_vision.py" --log "<LOG_LEVEL>" train \
    --train "/path/to/train_data" \
    --test "/path/to/test_data" \
    -s <split_train_val_ratio> -b <batch_size> --img-size "<width>x<height>" -e <num_epochs> \
    --model-checkpoint "/path/to/model_checkpoints/cv_model.{epoch:02d}-train-{accuracy:.4f}-{loss:.5f}-val-{val_accuracy:.4f}-{val_loss:.5f}.keras"
    --map "/path/to/model_classes.json" \
!python computer_vision.py --log "INFO" train --verbose 2 \
    --train "D:\\GitHub\winstars-ai-internship-test\\animals\\train" \
    --test "D:\\GitHub\winstars-ai-internship-test\\animals\\test"\
    -s 0.15 -b 64 --img-size 256 -e 2 \
    --model-checkpoint "./model_checkpoints/cv_model.{epoch:02d}-train-{accuracy:.4f}-{loss:.5f}-val-{val_accuracy:.4f}-{val_loss:.5f}.keras" \
    --map "./_model_classes.json"
Found 5356 files belonging to 11 classes.
Using 4553 files for training.
Found 5356 files belonging to 11 classes.
Using 803 files for validation.
Epoch 1/2
72/72 - 338s - 5s/step - accuracy: 0.3457 - loss: 1.9696 - val_accuracy: 0.9365 - val_loss: 0.7719 - learning_rate: 0.0010
Epoch 2/2
72/72 - 341s - 5s/step - accuracy: 0.6969 - loss: 1.1884 - val_accuracy: 0.9763 - val_loss: 0.3446 - learning_rate: 0.0010
Found 641 files belonging to 11 classes.

 1/11 ━━━━━━━━━━━━━━━━━━━━ 1:21 8s/step - accuracy: 0.9062 - loss: 0.8882
 2/11 ━━━━━━━━━━━━━━━━━━━━ 1:04 7s/step - accuracy: 0.9180 - loss: 0.7988
 3/11 ━━━━━━━━━━━━━━━━━━━━ 1:16 10s/step - accuracy: 0.9089 - loss: 0.7655
 4/11 ━━━━━━━━━━━━━━━━━━━━ 1:04 9s/step - accuracy: 0.9062 - loss: 0.7322 
 5/11 ━━━━━━━━━━━━━━━━━━━━ 54s 9s/step - accuracy: 0.9006 - loss: 0.7206 
 6/11 ━━━━━━━━━━━━━━━━━━━━ 40s 8s/step - accuracy: 0.8959 - loss: 0.7154
 7/11 ━━━━━━━━━━━━━━━━━━━━ 30s 8s/step - accuracy: 0.8945 - loss: 0.7061
 8/11 ━━━━━━━━━━━━━━━━━━━━ 20s 7s/step - accuracy: 0.8953 - loss: 0.6946
 9/11 ━━━━━━━━━━━━━━━━━━━━ 13s 7s/step - accuracy: 0.8967 - loss: 0.6822
10/11 ━━━━━━━━━━━━━━━━━━━━ 6s 6s/step - accuracy: 0.8987 - loss: 0.6690 
11/11 ━━━━━━━━━━━━━━━━━━━━ 0s 6s/step - accuracy: 0.9004 - loss: 0.6581
11/11 ━━━━━━━━━━━━━━━━━━━━ 64s 6s/step - accuracy: 0.9018 - loss: 0.6491

Predict Model
Example of predicting animal class on the image using the predict command:

python "./computer_vision.py" -l "<LOG_LEVEL>" predict -m "<MODEL_PATH>" --map "<CLASS_MAP_PATH>" -i "<IMAGE_PATH>"
!python ".\computer_vision.py" -l "DEBUG" predict \
    -m "D:\GitHub\winstars-ai-internship-test\task2-NER-image-classification\animals_model\model_checkpoints\cv_model.24-train-0.9212-0.29426-val-0.9950-0.03778.keras" \
    --map "_model_classes.json" \
    -i "..\\animals\\manual_testing\\1.lion.jpg"
1/1 ━━━━━━━━━━━━━━━━━━━━ 0s 4s/step
1/1 ━━━━━━━━━━━━━━━━━━━━ 4s 4s/step
lion

Examples of image classification
import os
import json
import logging
from types import SimpleNamespace

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from computer_vision import predict

# Folder containing the images
image_folder = r"D:\GitHub\winstars-ai-internship-test\animals\manual_testing"

# Model and map paths
model_path = r"D:\GitHub\winstars-ai-internship-test\task2-NER-image-classification\animals_model\model_checkpoints\cv_model.24-train-0.9212-0.29426-val-0.9950-0.03778.keras"
map_path = r"_model_classes.json"

# Get list of image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(("jpg", "jpeg", "png", "webp"))]

# Load class map for predicted class reference (if needed)
with open(map_path, "r") as file:
    class_map = json.load(file)

# Set up a logger (if needed)
logger = logging.getLogger(__name__)
logger.setLevel(logging.ERROR)


# Function to call your predict function
def get_prediction(image_path):
    # Arguments for the predict function
    args_dict = {"model_path": model_path, "class_mapping_file": map_path, "image_path": image_path}
    args = SimpleNamespace(**args_dict)

    # Call your predict function
    predicted_class = predict(args, logger=logger)
    return predicted_class


# Prepare to plot the images and predicted classes
cols = 5
rows = len(image_files) // cols + (1 if len(image_files) % cols != 0 else 0)  # Adjust the number of rows if needed
fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 5))

# Flatten the axes array for easier iteration
axes = axes.flatten()


for ax, image_file in zip(axes, image_files):
    image_path = os.path.join(image_folder, image_file)

    # Get the prediction for the current image
    predicted_class = get_prediction(image_path)

    # Plot the image
    img = mpimg.imread(image_path)
    ax.imshow(img)

    # Extract the true class from the image filename (e.g., '1.bear.jpg' -> 'bear')
    true_class = image_file.split('.')[1]
    # Compare true class and predicted class
    title_color = 'green' if true_class.lower() == predicted_class.lower() else 'red'
    # Set the title with the appropriate color
    ax.set_title(f"True:{true_class} - Predicted: {predicted_class}", color=title_color)
    ax.axis("off")  # Hide the axes

plt.tight_layout()
plt.show()
```
